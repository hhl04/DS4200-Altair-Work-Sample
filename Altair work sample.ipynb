##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
```

```python
question_1 = alt.Chart(weather).mark_boxplot().encode(
    #weather is nominal
    alt.X('weather:N', title='Weather Type'),
    #temp_max is quantitative data with magnitude
    alt.Y('temp_max:Q', title='Max Temperature'),

    alt.Color('weather:N', title = 'Weather Type'),
    alt.Tooltip(['weather:N', 'temp_max:Q'])
).properties(
    title='Max Temperature by Weather Type',
    width=300,
    height=200
)
question_1
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
question_2 = alt.Chart(weather).mark_circle().encode(
    #temp_max is quantitative data with magnitude
    alt.X('temp_max:Q', title='Max Temperature'),
    #temp_min is quantitative data with magnitude
    alt.Y('temp_min:Q', title='Min Temperature'),

    alt.Color('weather:N', title = 'Weather Type'),
    alt.Tooltip(['weather:N', 'temp_max:Q', 'temp_min:Q'])
).properties(
    title='Max Temperature vs Min Temperature with Weather Type'
)

question_2
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels.

```python
question_3 = alt.Chart(weather).mark_circle().encode(
    alt.X('temp_max:Q', title='Max Temperature'),
    alt.Y('temp_min:Q', title='Min Temperature'),

    #wind speed is quantitative
    alt.Color('wind:Q', title = 'Wind Speed',
              #darker color = faster wind
              scale=alt.Scale(range=['orange', 'blue'])),

    alt.Tooltip(['weather:N', 'temp_max:Q', 'temp_min:Q', 'wind:Q'])
).facet(
    #what attributes to focus the graph on
    #define the layout of the facet
    facet = 'weather:N',
    columns = 3
)
question_3
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python
part_1 = alt.Chart(weather).mark_line().transform_calculate(
    avg_temp = '(datum.temp_min + datum.temp_max) / 2'
).encode(
    alt.X('date:T', title='Date'),
    alt.Y('avg_temp:Q', title='Average Temperature')
).properties(
    title='Average Temperature Over Time',
    width = 300,
    height = 200
)

part_2 = alt.Chart(weather).mark_circle().transform_calculate(
    avg_temp = '(datum.temp_min + datum.temp_max) / 2',

).encode(
    alt.X('date:T', title='Date'),
    alt.Y('avg_temp:Q', title='Average Temperature'),
    alt.Color('year(date):T', scale=alt.Scale(
                              range=['crimson', 'yellow', 'palegreen',
                                     'lightblue'])
              )
).transform_filter(
    #this highlights the highest average temperature
    'datum.avg_temp >= 25'
).properties(
    title='Highest Average Temperature Over Time',
    width = 300,
    height = 200
)
part_1 + part_2
```

##### Question 1 (5 points)

Following is a density plot for the wind speed for all five types of weather. Is there any problems in the figure? Try to name at least two ways you can do to re-design the figure (change the mark/channel/encoding etc.)

```python
# Create density plots using Altair
density_plot = alt.Chart(weather).transform_density(
    'wind',
    groupby=['weather'],
    as_=['wind', 'density']
).mark_area(opacity=0.7).encode(
    #change OPACITY
    alt.X('wind:Q'),
    alt.Y('density:Q'),
    alt.Color('weather:N', scale=alt.Scale(
            range=['lightgray', 'palegreen', 'skyblue', 'lightcoral', 'crimson']))
).properties(
    title='Density Plot of Wind for Sun, Drizzle, and Snow'
)

# Show the plot
density_plot.display()
```

We can change the opacity of each attribute, or switch the order of colors. <br>
We can create a facet figure of 2 rows and 3 columns, and display the density plots separately. <br>

People having `Anomalous Trichromacy` can see all RGB shades, but one of these RGB elements may be a little bit off for them. These patients may not see a lighter shade of red or blue or green. If these patients can't see as much red or green, my picture appears a bit darker for them. If these patients can't see as much blue, my picture appears brighter for them.<br>

`Dichromatic view` means the patients can only see 2 out of 3 colors in the RGB range. If these patients can't see the red color, my picture appears more yellow. If these patients can't see the green color, my picture appears more orange. If these patients can't see the blue color, my picture appears more *blue*, which i find interesting. <br>

`Monochromatic view` means the patients can only see black and white, or only light green and grey.
